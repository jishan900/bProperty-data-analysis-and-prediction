```python
n_df = df[["beds", "bath","type", "Location", "area_sqft", "Price"]]
n_df.head(3)
```

```python
n_df = pd.get_dummies(n_df, drop_first = True)
n_df.head()
```

```python
x = n_df.drop("Price", axis = 1)
y = n_df["Price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x,y, test_size=0.2)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(x,y)
```

```python
y_pred = clf.predict(x_train)
y_pred[:50]
```

```python
y_pred = clf.predict(x_test)
y_pred[:50]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from math import sqrt
msr = mean_squared_error(y_test, y_pred)
print('Mean squared error (MSE):', msr)
rs = r2_score(y_test, y_pred)
print('R2 score:', rs)
mae = mean_absolute_error(y_test, y_pred)
print('Mean absolute error (MAE):', mae)
RMSE = sqrt(mean_squared_error(y_test, y_pred))
print('Root Mean Square Error (RMSE):', RMSE)
msr, rs, mae, RMSE
```

XGBRegressor

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(x,y)
xgb_pred = xgb.predict(x_test)
xgb_msr = mean_squared_error(y_test, xgb_pred)
xgb_rs = r2_score(y_test, xgb_pred)
xgb_mae = mean_absolute_error(y_test, xgb_pred)
xgb_RMSE = sqrt(mean_squared_error(y_test, y_pred))
print("Mean Squared Error (MSE): ", xgb_msr)
print("R2 score: ", xgb_rs)
print("Mean Absolute Error (MAE): ", xgb_mae)
print('Root Mean Square Error (RMSE):', RMSE)
```

LGBMRegressor

```python
from lightgbm import LGBMRegressor
lgbm = LGBMRegressor()
lgbm.fit(x,y)
```

```python
lgbm_pred = lgbm.predict(x_test)
lgbm_msr = mean_squared_error(y_test, lgbm_pred)
lgbm_rs = r2_score(y_test, lgbm_pred)
lgbm_mae = mean_absolute_error(y_test, lgbm_pred)
lgbm_RMSE = sqrt(mean_squared_error(y_test, y_pred))
print("Mean Squared Error (MSE): ", lgbm_msr)
print("R2 score: ", lgbm_rs)
print("Mean Absolute Error (MAE): ", lgbm_mae)
print('Root Mean Square Error (RMSE):', lgbm_RMSE)
```
